## Investigation Overview

> The goal of this presentation is to show: 
> 1. What time of days are most popular for bike rentals
> 2. How bike rental duration changes with age
> 3. How the average bike rental duration changes for each day of the week, for each user type (subscriber vs. customer)

## Dataset Overview

> This data set included information on the trips taken for the Ford GoBike system in San Franciso in February - March 2019. The dataset includes information about the user's ages, gender, trip duration, start and end destination, time of day trips were taken and day of week the trips were taken. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('df_clean.csv')
```

## Frequency of bike rentals for each hour of the day

> Most bike rentals occur in the mornings (around 8am) and in the evenings (around 6pm). These two peaks align with the start and end of the normal 9-5 workday. 

```python
a = sb.displot(df_clean['start_time_hour'], kde=True, height = 7, aspect = .99)
a.set_axis_labels("Hour", "Frequency of bike rentals)")
plt.title("Frequency of bike rentals for each hour of the day")
ticks = list(range(24))
plt.xticks(ticks[::1])
```

## Duration of bike rentals for different ages

> For all ages, most rentals occur for less than ± 3 hours (180 minutes). Younger ages (<40 years) tend to rent bikes for longer duration than older ages (>40 years). 

```python
plt.scatter(data=df_clean, x='duration_minute' ,y='age', alpha=5/20);
plt.xlabel('duration (minutes)')
plt.ylabel('ages');
plt.title('Duration (minutes) of bike rentals for different ages')
```

## Average trip duration per day for each user type
> Customers tend to have longer trip durations throughout the week than subscribers. There is a peak in average duration of bike rentals on the weekends for customers.

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sb.pointplot(data = df_clean, x = 'start_time_weekday', y = 'duration_minute', order=weekday, hue='user_type', dodge=0.2, linestyles="");
plt.xlabel('Day of the week');
plt.ylabel('Average trip duration in minutes');
plt.legend(title='User type')
plt.title('Average duration (minutes) of bike rentals for different days of the week, comparing user types')
```

## Overall conclusion

Bike rentals peak in the morning and evenings, most rentals are undertaken for less than 3 hours, with older people renting for shorter durations, and customers rent for longer periods each day of the week. Rentals are higher on weekends than weekdays.
